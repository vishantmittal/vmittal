```python
message = "   Hello World Python   "

# Convert to uppercase
print(f"Uppercase: {message.upper()}")

# Replace a substring
print(f"Replace 'Python' with 'Colab': {message.replace('Python', 'Colab')}")

# Split the string into a list of words
print(f"Split by space: {message.split()}")

# Remove leading/trailing whitespace
print(f"Stripped: '{message.strip()}'")

# Check if string starts or ends with a prefix/suffix
print(f"Starts with '   Hello': {message.startswith('   Hello')}")
print(f"Ends with 'Python   ': {message.endswith('Python   ')}")
```

```python
first_name = "John"
last_name = "Doe"
age = 30

# Concatenating strings using f-string
full_greeting = f"Hello, my name is {first_name} {last_name} and I am {age} years old."
print(full_greeting)

# You can also include expressions directly
product = 5
price = 10.50
total_cost = f"The total cost for {product} items is ${product * price:.2f}."
print(total_cost)
```

Here's a brief explanation of the methods used:

*   **`message.upper()`**: Returns a copy of the string with all characters converted to uppercase.
*   **`message.replace('old', 'new')`**: Returns a copy of the string with all occurrences of substring `old` replaced by `new`.
*   **`message.split()`**: Splits the string by whitespace (or a specified delimiter) and returns a list of substrings.
*   **`message.strip()`**: Returns a copy of the string with leading and trailing whitespace removed.
*   **`message.startswith('prefix')`**: Returns `True` if the string starts with the specified `prefix`, otherwise `False`.
*   **`message.endswith('suffix')`**: Returns `True` if the string ends with the specified `suffix`, otherwise `False`.
